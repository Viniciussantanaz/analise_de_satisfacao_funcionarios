```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('employee_survey.csv')
```

# Descrição das Colunas

| Nome da Coluna               | Descrição                                                                                       |
|------------------------------|-------------------------------------------------------------------------------------------------|
| **EmpID**                    | Identificador único para cada funcionário.                                                     |
| **Gender**                   | Gênero do funcionário (por exemplo, masculino, feminino, outro).                               |
| **Age**                      | Idade do funcionário.                                                                           |
| **MaritalStatus**            | Estado civil do funcionário (por exemplo, solteiro, casado, divorciado, viúvo).              |
| **JobLevel**                 | Nível de cargo do funcionário (por exemplo, estagiário/calouro, júnior, intermediário, sênior, líder). |
| **Experience**               | Número de anos de experiência de trabalho que o funcionário possui.                             |
| **Dept**                     | Departamento onde o funcionário trabalha (por exemplo, TI, RH, Finanças, Marketing, Vendas, Jurídico, Operações, Atendimento ao Cliente). |
| **EmpType**                  | Tipo de emprego (por exemplo, período integral, meio período, contrato).                       |
| **WLB**                      | Avaliação do equilíbrio entre vida pessoal e profissional (escala de 1 a 5).                   |
| **WorkEnv**                  | Classificação do ambiente de trabalho (escala de 1 a 5).                                       |
| **PhysicalActivityHours**     | Número de horas de atividade física por semana.                                               |
| **Workload**                 | Classificação da carga de trabalho (escala de 1 a 5).                                         |
| **Stress**                   | Classificação do nível de estresse (escala de 1 a 5).                                          |
| **SleepHours**               | Número de horas de sono por noite.                                                             |
| **CommuteMode**              | Modo de deslocamento (por exemplo, carro, transporte público, bicicleta, caminhada, motocicleta). |
| **CommuteDistance**          | Distância percorrida durante o trajeto (em quilômetros).                                       |
| **NumCompanies**             | Número de empresas diferentes para as quais o funcionário trabalhou.                           |
| **TeamSize**                 | Tamanho da equipe da qual o funcionário faz parte.                                            |
| **NumReports**               | Número de pessoas às quais o funcionário se reporta (aplicável somente aos níveis Sênior e Líder). |
| **EduLevel**                 | Maior nível de educação alcançado pelo funcionário (por exemplo, ensino médio, bacharelado, mestrado, doutorado). |
| **haveOT**                   | Indicador se o funcionário tem horas extras (Verdadeiro/Falso).                                 |
| **TrainingHoursPerYear**     | Número de horas de treinamento recebidas por ano.                                             |
| **JobSatisfaction**          | Classificação da satisfação no trabalho (escala de 1 a 5).                                     |

```python
satisfacao_funcionarios = pd.DataFrame(df.JobSatisfaction.value_counts(normalize=True))
satisfacao_funcionarios.reset_index(inplace=True)
satisfacao_funcionarios.JobSatisfaction = satisfacao_funcionarios.JobSatisfaction.astype(str)
satisfacao_funcionarios = satisfacao_funcionarios.sort_values(by='JobSatisfaction').reset_index(drop=True)
satisfacao_funcionarios
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,5))

ax.plot(satisfacao_funcionarios.JobSatisfaction, satisfacao_funcionarios.proportion, alpha=0.8, color='#41546D', marker='o',  label='Distribuição das notas de satisfação')
for spine in ax.spines.values():
    spine.set_visible(False)

plt.legend(loc='upper left', fontsize=13)
plt.show()
```

<p style= "font-size:16px; background-color:'gray'; font-family: 'Helvetica'"><b>aprox.58% dos funcionários</b> (aqueles com a satisfação entre 4 e 5) se <b>mostrou está muito ou pouco satisfeito com empresa</b>.<br />
 <b>aprox.18% dos funcionários está neutro, não está satisfeito e também não está insatisfeito com a empresa</b>(aqueles com a satifação igual a 3).<br /> <b>aprox.13% dos funcionários está muito insatifeito com a empresa (aqueles com a satisfação igual a 1). </b> Ao todo <b>aprox.22% dos funcionários (aqueles com a satifação entre 1 e 2)</b> está pouco ou muito insatisfeito com a empresa, vamos entender o porquê 22% dos funcionários está insatisfeito. </p>

```python
funcionarios_insatisfeito = df[df.JobSatisfaction < 3]
funcionarios_satisfeito =  df[df.JobSatisfaction > 3]
```

```python
dept_insatisfeito = pd.DataFrame(funcionarios_insatisfeito.Dept.value_counts(normalize=True)).rename(columns={'proportion':'Funcionários insatisfeito'})
dept_satisfeito = pd.DataFrame(funcionarios_satisfeito.Dept.value_counts(normalize=True)).rename(columns={'proportion':'Funcionários satisfeito'})
departamentos = pd.concat([dept_insatisfeito,dept_satisfeito], axis=1).reset_index()
departamentos.sort_values(by='Funcionários satisfeito', ascending=True, inplace = True)
departamentos.reset_index(inplace=True, drop=True)
departamentos
```

## Setores com funcinários mais insatisfeito e satisfeitos

```python
labels = list(departamentos.Dept)
valores_1 = list(departamentos['Funcionários insatisfeito'])
valores_2 = list(departamentos['Funcionários satisfeito'])

x = np.arange(len(labels))  
largura = 0.35  

fig, ax = plt.subplots(1,1, figsize=(20,5))

color1 = ['#87A8DE','#7692C1','#647CA5','#536789','#42526D','#313D51','#1F2834','#1F2834']
color2 = ['#246FDF','#1F60C1','#1A52A5','#164489','#11366D','#0D2851','#081A34','#081A34'] 

barras1 = ax.bar(x - largura/2, valores_1, largura, label='Funcionários insatisfeito',color=color2)

barras2 = ax.bar(x + largura/2, valores_2, largura, label='Funcionários satisfeito', color=color1)
for spine in ax.spines.values():
    spine.set_visible(False)

ax.set_xlabel('Departamentos')
ax.set_ylabel('Porcentagem')
ax.set_title('Porcentagem de satisfação e insatisfação dos funcionários ',  color='#666666',fontsize=14, loc='left')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

<P style= "font-size:16px; background-color:'gray'; font-family: 'Helvetica'"> O que podemos concluir é que não existe departamnetos 'bons' ou 'ruins', a insatifação e satisfação dos funcionários está bem divida em cada departamento  da empresa. No departamento do TI por exemplo, dos 57% (1765) funcionários satisfeito, 24% é do departamento do TI, e dos 22% (694) funcionários insatisfeitos, 28% é do TI, e esse equilíbrio acontece em todos os setores da empresa. Ou seja, talvez para alguns funcionários os departamentos não seja o maior responsável pela insatisfação.  </P>

## Correlações

### Correlaçãos entre as notas de satisfação dos funcionários insatisfeito e satisfeito com as outras variáveis.

```python
columns = df[['JobSatisfaction','TrainingHoursPerYear', 'TeamSize', 'NumCompanies', 'CommuteDistance', 'Experience','Age','PhysicalActivityHours']]


correlation = columns.corr()

plt.figure(figsize=(20,5))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

plt.show()
```

### Correlaçãos mais fortes entre as notas de satisfação dos funcionários insatisfeito e satisfeito com as outras variáveis.

```python
columns = funcionarios_satisfeito[['SleepHours','WLB','WorkEnv','JobSatisfaction','Stress','Workload']]

correlation = columns.corr()

plt.figure(figsize=(20,5))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

plt.show()
```

```python
columns = funcionarios_insatisfeito[['SleepHours','WLB','WorkEnv','JobSatisfaction','Stress','Workload']]


correlation = columns.corr()

plt.figure(figsize=(20,5))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

plt.show()
```

<p style= "font-size:16px; background-color:'gray'; font-family: 'Helvetica'"> As variáveis (TrainingHoursPerYear, TeamSize, NumCompanies, CommuteDistance, Experience, Age, PhysicalActivityHours) não tiveram nenhuma correlção com a nota de satisfação dos funcionários, já as variáveis <b>SleepHours (Horas de sono)</b> <b> WLB (equilíbrio entre vida pessoal e profissional)</b>, <b> WorkEnv (ambiente de trabalho)</b> <b> Stress (nível de estresse)</b> <b> Workload (carga horária de trabalho) tiveram uma correlação consideravalmente boa com as notas de classificação dos funcinários insatisfeios, com exceção do <b>SleepHours</b> e do<b> Stress</b>, as otras tiveram uma correlção acima de 0.32, ainda é uma correlação fraca, mas não deixa de ser interessante, o curioso é que essas mesmas variáeis não tiveram a mesma correlação nas notas de satisfação dos funcionários satisfeios, e até por isso não dá para dizer se esses indicadores são uma das causas da insatisfação.</p

## Comparações

```python

stress = pd.concat([pd.DataFrame(funcionarios_satisfeito.Stress.value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_satisfeito'}),
                pd.DataFrame(funcionarios_insatisfeito.Stress.value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_insatisfeito'})], axis=1)

wlb = pd.concat([pd.DataFrame(funcionarios_satisfeito.WLB.value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_satisfeito'}),
                pd.DataFrame(funcionarios_insatisfeito.WLB.value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_insatisfeito'})], axis=1)

WorkEnv = pd.concat([pd.DataFrame(funcionarios_satisfeito.WorkEnv.value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_satisfeito'}),
                pd.DataFrame(funcionarios_insatisfeito.WorkEnv.value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_insatisfeito'})], axis=1)

Workload = pd.concat([pd.DataFrame(funcionarios_satisfeito.Workload.value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_satisfeito'}),
                pd.DataFrame(funcionarios_insatisfeito.Workload.value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_insatisfeito'})], axis=1)
```

```python
media_de_sono = pd.DataFrame({'Sono Func. Satisfeito': [funcionarios_satisfeito.SleepHours.mean().round(2)], 'Sono Func. Isatisfeito': [funcionarios_insatisfeito.SleepHours.mean().round(2)]})
media_de_sono.rename(index={0:'Média'}, inplace= True)
```

```python
stress = stress.reset_index()
wlb = wlb.sort_index()
wlb = wlb.reset_index()
WorkEnv = WorkEnv.sort_index()
WorkEnv = WorkEnv.reset_index()
Workload = Workload.sort_index()
Workload = Workload.reset_index()
```

```python
fig, ax = plt.subplots(1,4, figsize=(20,5))

var = [stress, wlb, WorkEnv, Workload]
titles = ['Nível de Estresse', 'equilíbrio entre vida pessoal e profissional', 'ambiente de trabalho', 'carga horária de trabalho']

for i in range(len(var)):
    labels = list(['1','2','3','4','5'])
    valores_1 = list(var[i]['funcionarios_insatisfeito'])
    valores_2 = list(var[i]['funcionarios_satisfeito'])

    x = np.arange(len(labels))  
    largura = 0.35  



    barras1 = ax[i].bar(x - largura/2, valores_1, largura, label='Funcionários insatisfeito', color='#405069')

    barras2 = ax[i].bar(x + largura/2, valores_2, largura, label='Funcionários satisfeito',color='#113469')
    ax[i].set_xticks(x)
    ax[i].set_xticklabels(labels)
    ax[i].set_title(titles[i], color='#666666',fontsize=12)
    for spine in ax[i].spines.values():
        spine.set_visible(False)

ax[0].set_ylabel('Porcentagem')
ax[0].legend()

plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(20,5))

ax[0].hist(funcionarios_insatisfeito.SleepHours, bins=12, color='#536B8A')
ax[0].set_title('Distribuição do sono dos funcionários insatisfeito')
ax[0].set_xlabel('Horas')
ax[0].set_ylabel('Qtd. Funcionários')
for spine in ax[0].spines.values():
    spine.set_visible(False)

ax[1].hist(funcionarios_satisfeito.SleepHours, bins=12, color='#41546D')
ax[1].set_title('Distribuição do sono dos funcionários satisfeitos')
ax[1].set_xlabel('Horas')
for spine in ax[1].spines.values():
    spine.set_visible(False)
plt.show()
```

Olhando somente para as notas 1 e 5 percebemos que sempre há uma descrpância dos funcionários satisfeitos e insatifeitos, as variáveis WLB (equilíbrio entre vida pessoal e profissional) e WorkEnv (ambiente de trabalho) por exemplo, mais de 25% dos funcionários que classificaram a nota de satisfação 1 ou 2 não conseguem manter o quilíbrio entre vida pessoal e profissional ou acham o ambiente de trabalho ruim, do outro lado, mais 25% dos dos funcionários que classificaram a nota de satisfação 4 ou 5 tem essas duas coisas muito bem equilíbrada.  </p>

## Comparando outras variáveis

```python
fig, ax = plt.subplots(1,5, figsize=(20,5))
var = ['EduLevel','CommuteMode','EmpType','JobLevel']

for i in range(len(var)):
    columns1 = pd.DataFrame(funcionarios_satisfeito[var[i]].value_counts(normalize=True)).sort_values(by='proportion', ascending=True).reset_index()

    x = list(columns1[var[i]])
    y1 = columns1.proportion
    ax[i].plot(x,y1, label='Satisfeitos')
    
    columns2 = pd.DataFrame(funcionarios_insatisfeito[var[i]].value_counts(normalize=True)).sort_values(by='proportion', ascending=True).reset_index()
    y2 = columns2.proportion
    ax[i].plot(x,y2, label='Insaifeitos')
    ax[i].set_title(var[i])

    ax[i].set_xticks(range(len(x)))
    ax[i].set_xticklabels(x, rotation=45)

MaritalStatus = pd.concat( [ pd.DataFrame(funcionarios_insatisfeito[['MaritalStatus']].value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_insatisfeito'}), 
            pd.DataFrame(funcionarios_satisfeito[['MaritalStatus']].value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_satisfeito'}) ], axis=1 )
MaritalStatus.fillna(0, inplace=True)
MaritalStatus = MaritalStatus.sort_values(by='funcionarios_satisfeito', ascending=True).reset_index()

x = MaritalStatus.MaritalStatus
y1 = MaritalStatus.funcionarios_satisfeito
y2 = MaritalStatus.funcionarios_insatisfeito
ax[4].plot(x,y1)
ax[4].plot(x,y2)
ax[4].set_title('MaritalStatus')
ax[4].set_xticks(range(len(x)))
ax[4].set_xticklabels(x, rotation=45)

ax[0].legend()
plt.show()
```

<p style= "font-size:16px; background-color:'gray'; font-family: 'Helvetica'"> Os gráficos mmostram um comparativo das variáveis, <b>EduLevel</b> (Nível de escolaridade), <b>CommuteMode</b> (Meio de transporte), <b>EmpType</b> (Tipo de contrato), <b>JobLevel</b> (Senioridade), entre os insatisfeitos e satisfeitos, o objetio é tentar ver se os funcionários satisfeitos têm algo possitivo a mais ou negativo a menos em relação aos funcionários insatisfeitos, mas os resultados são muito iguais.</p>

```python
pd.concat( [ pd.DataFrame(funcionarios_insatisfeito[['JobLevel']].value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_insatisfeito'}), 
            pd.DataFrame(funcionarios_satisfeito[['JobLevel']].value_counts(normalize=True)).rename(columns={'proportion':'funcionarios_satisfeito'}) ], axis=1 ).sort_values(by='funcionarios_satisfeito', ascending=False).reset_index()
```

```python
senior_insatisfeito = funcionarios_satisfeito[(funcionarios_satisfeito.JobLevel == 'Senior') | (funcionarios_satisfeito.JobLevel == 'Lead') ]
senior_satisfeito = funcionarios_insatisfeito[(funcionarios_insatisfeito.JobLevel == 'Senior') | (funcionarios_insatisfeito.JobLevel == 'Lead') ]
```

```python
df_senior_lider= pd.concat([senior_insatisfeito, senior_satisfeito])
```

```python
columns = df_senior_lider[['JobSatisfaction','NumReports']]


correlation = columns.corr()

plt.figure(figsize=(20,5))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

plt.show()
```

<p style= "font-size:16px; background-color:'gray'; font-family: 'Helvetica'">Como o gráfico indicava, a tabela mostra a distribuição da senioridade entre os funcinários insatisfeito e satisfeito, e os seniors ocupam 35% em cada grupo  e os líderes ocupam 13% / 12% em cada grupo, e a variável NumReports (Número de pessoas às quais o funcionário se reporta (aplicável somente aos níveis Sênior e Líder)). poderia ser umas das causas da insatisfação desses funcionários, mas como mostra o gráfico não existe correlção enrtre essas duas variáveis.)
</p>

## Conclusão 

<p style= "font-size:16px; background-color:'gray'; font-family: 'Helvetica'">Apontar a principal causa da insatisfação dos funcionários é algo muito complexo, um caso de endogeneidade, as causas da insatisfação pode ser carga horária de trabalho, nível de estress, ambiente de trabalho, equilíbrio entre vida pessoal e profissional, Horas de sono, ou outro problema pessoal. Mas como os dados apontam uma das causas pode ser um desses, talvez se empresa melhorar nas partes onde tem controle como carga horária de trabalho e ambiente de trabalho, a insatisfação desses funcinários diminua.   </p>
